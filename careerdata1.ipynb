```python

    from google.colab import drive

    drive.mount('/content/drive')
    file_path = '/content/drive/MyDrive/career_data1.csv'

    import pandas as pd

    df = pd.read_csv(file_path)
    df.head()
```

```python
# Convert categorical columns to lowercase for consistency
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
#One-hot encode multiple-choice responses (e.g., AI Platform Interest)
df['AI Platform Interest'] = df['AI Platform Interest'].map({'yes': 1, 'maybe': 0.5, 'no': 0})

# Convert 'Confidence' to numerical
df['Confidence'] = df['Confidence'].map({'yes': 1, 'no': 0})

# Expand 'Challenges' column into multiple binary columns
challenges_list = ["Lack of career guidance", "Limited exposure to career options", "Financial limitations"]
for challenge in challenges_list:
    df[challenge] = df['Challenges'].apply(lambda x: 1 if challenge.lower() in str(x).lower() else 0)

# Drop the original text-based 'Challenges' column
df.drop(columns=['Challenges'], inplace=True)

# Save cleaned data
df.to_csv("career_data_cleaned.csv", index=False)

print("Data cleaning complete! Ready for ML.")
```
